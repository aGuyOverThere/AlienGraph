```python
import requests
from jinja2 import Environment, FileSystemLoader
import pandas as pd
from datetime import datetime as dt
import re
```

```python
uri = "https://en.wikipedia.org/wiki/List_of_reported_UFO_sightings"
rep_rex = re.compile(r'\[\d{1,5}\]')
```

```python
df = pd.read_html(uri)
```

```python
df[0:6]
n_df = pd.concat(df[0:7])
n_df.dropna(subset=['Name'], inplace=True)
n_df.drop(columns=['.mw-parser-output .tooltip-dotted{border-bottom:1px dotted;cursor:help}Date'], inplace=True)
```

```python
n_df
```

```python
environment = Environment(loader=FileSystemLoader("./_templates/"))
template = environment.get_template("ðŸ”¥ Incident Template.md")
```

```python
for index, row in n_df.iterrows():
    f_name = f"../ðŸ”¥ Incidents/ðŸ”¥ {row['Name']}.md"
    date = dt.now()
    data = {
        "author": "aGuyOverThere", 
        "source": f"[Wikipedia Ufo Sightings]({uri})",
        "type": "#sighting",
        "incident_date": row['Date'],
        "reported_date": "unknown",
        "incident_author": "Wikipedia",
        "incident_location": row['Location'],
        "incident_summary": rep_rex.sub("", row['Description']),
        "created_date": date.strftime('%m/%d/%Y'),
        "created_year": date.year,
        "mm_yyyy": date.strftime('%m_%Y'),
        "id": row['Name'].split()[0]
    }
    content = template.render(data)
    with open(f_name, mode="w", encoding="utf-8") as message:
        message.write(content)
    
```
